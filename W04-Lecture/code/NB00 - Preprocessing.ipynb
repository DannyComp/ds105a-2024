```python
import os
```

```python
data_folder = os.path.join('..', 'data', 'waitrose-2024-07')
```

```python
# Use a list comprehension to get all the files in the folder
all_files = [os.path.join(data_folder, file) for file in os.listdir(data_folder) 
             if file.endswith('.csv')]
```

```python
import pandas as pd
```

```python
# Read every single file as a DataFrame
# Save the dataframes in a list
list_of_dfs = [pd.read_csv(file) for file in all_files]
```

```python
# Use pd.concat to concatenate all the files into a single DataFrame
df = pd.concat(list_of_dfs)

# Check that we have all the data
df.info()
```

```python
def clean_item_price(item_price: str) -> float: 
    if '£' in item_price:
        # Remove the '£' symbol
        item_price = item_price.replace('£', '')
    if 'p' in item_price:
        # Remove the '£' symbol
        cleaned_price = float(item_price.replace('p', ''))/100
    # Remove commas and strip whitespace, then convert to float
    cleaned_price = float(item_price.replace(',', '').strip())
    return cleaned_price

Example = '75p'
cleaned_price = clean_item_price(Example)
print(cleaned_price)
type(cleaned_price)
```
